```python

print("="*50)
print("Importing libraries...")
print("="*50)

import pandas as pd
import numpy  as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
print("="*50)
print("Loading dataset...")
print("="*50)
df = pd.read_csv("../Task_2/data/titanic_cleaned.csv")
```

```python
print("="*50)
print("Dataset information:")
print("="*50)
df.info()
```

```python
df.head(5)
```

```python
# Statistical summary of the dataset
df.describe()
```

```python
print("="*50)
print("Average fare by passenger class:")
print("="*50)
avg_fare_by_class = df.groupby('Pclass')['Fare'].mean().sort_values(ascending=False)
avg_fare_by_class
```

```python
print("="*50)
print("Overall fare statistics:")
print("="*50)
df['Fare'].mean()
df['Fare'].max()
df['Fare'].min()
```

```python
# Distribution of Age
print("Distribution of Age:")
print("="*50)

plt.hist(df['Age'],bins=20)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')
plt.show()
```

```python
print("Boxplot of Age:")

plt.boxplot(df['Age'])

plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')
plt.show()
```

```python
print("Boxplot of Age by Survival Status:")

sns.boxplot(x='Survived', y='Age', data=df)
plt.xlabel('Survived')
plt.ylabel('Age')
plt.title('Boxplot of Age by Survival Status')
plt.show()
```

```python
print("Scatter plot of Age vs Fare:")
sns.scatterplot(x= 'Age',y='Fare',data=df)
plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Age vs Fare")
plt.show()
```

```python
corr = df.corr(numeric_only=True)
print("Correlation matrix:")

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
print("Bar plot of Survival rate by Sex:")
sns.barplot(x='Sex', y='Survived', data=df)
plt.xlabel('Sex')
plt.ylabel('Survival Rate')
plt.title('Bar plot of Survival rate by Sex')
plt.show()

print("Bar plot of Survival rate by Passenger Class:")
sns.barplot(x='Pclass', y='Survived', data=df)
```
